```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
from sklearn import datasets
from sklearn.metrics import silhouette_score


iris = datasets.load_iris()
X = iris.data

# Number of samples,clusters
k = 3

kmeans = KMeans(n_clusters=k, random_state=0)
kmeans_labels = kmeans.fit_predict(X)

# Silhouette Score for K-Means
print("K-Means Silhouette Score:", silhouette_score(X, kmeans_labels, metric='euclidean'))

kmedoids = KMedoids(n_clusters=k, random_state=0)
kmedoids_labels = kmedoids.fit_predict(X)

# Silhouette Score for K-Medoids
print("K-Medoids Silhouette Score:", silhouette_score(X, kmedoids_labels, metric='euclidean'))

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels)
plt.title("K-Means Clustering")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.subplot(1,2,2)
plt.scatter(X[:, 0], X[:, 1], c=kmedoids_labels)
plt.title("K-Medoids Clustering")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.show()
```
