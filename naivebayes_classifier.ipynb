```python
# Import library
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler, Binarizer

# Load data
data = load_breast_cancer()
X = data.data
y = data.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Gaussian Naive Bayes
g = GaussianNB()
g.fit(X_train, y_train)
pred1 = g.predict(X_test)
print("Gaussian:", accuracy_score(y_test, pred1))

# Multinomial Naive Bayes
scaler = MinMaxScaler()
X_train_m = scaler.fit_transform(X_train)
X_test_m = scaler.transform(X_test)

m = MultinomialNB()
m.fit(X_train_m, y_train)
pred2 = m.predict(X_test_m)
print("Multinomial:", accuracy_score(y_test, pred2))


# Bernoulli Naive Bayes
binarizer = Binarizer()
X_train_b = binarizer.fit_transform(X_train)
X_test_b = binarizer.transform(X_test)

b = BernoulliNB()
b.fit(X_train_b, y_train)
pred3 = b.predict(X_test_b)
print("Bernoulli:", accuracy_score(y_test, pred3))
```
