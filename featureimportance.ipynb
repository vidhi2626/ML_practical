```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("auto-mpg.csv")   # use your actual file name

#  Check columns (IMPORTANT)
print("Columns in dataset:", df.columns)

# Separate features and target
X = df.drop('mpg', axis=1)   # mpg is target in auto-mpg dataset
Y = df['mpg']

#  REMOVE ALL STRING COLUMNS (BEST FIX)
X = X.select_dtypes(exclude=['object'])

# OR (if you want to keep categorical data, use this instead)
# X = pd.get_dummies(X, drop_first=True)

# Split data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train model
from sklearn.tree import DecisionTreeRegressor
clf = DecisionTreeRegressor(random_state=42)
clf.fit(X_train, Y_train)

print("Model trained successfully ")



```
