```python
#PCA on auto MPG dataset
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#Load the datset
df=pd.read_csv("C:/Users/Admin/Desktop/ml_practical/auto-mpg.csv")

#Data Preprocessing
#replace ? with Nan
df.replace('?',np.nan,inplace=True)

#convert horsepower to numeric
df['horsepower']=pd.to_numeric(df['horsepower'])

#drop rows with missing values
df.dropna(inplace=True)

#Drop non-numeric coumn(car name)
if 'car name' in df.columns:
    df.drop(columns=['car name'],inplace=True)

#feature selection
#separate features and targer (mpg)
X=df.drop(columns=['mpg'])
Y=df['mpg']

#Convert categorical column 'origin' using one-hot encoding
X=pd.get_dummies(X,columns=['origin'],drop_first=True)

#Feature scaling (important for PCA)
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

#Apply PCA
#example : reduce to 3 principal components
pca=PCA(n_components=3)
X_pca=pca.fit_transform(X_scaled)

#Results
print("Original Shape:",X_scaled.shape)
print("Reduce Shape:",X_pca.shape)
```
