```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# sample data (internal vs external marks)
internal = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
external = np.array([15, 25, 35, 45, 55])

# create model
model = LinearRegression()
model.fit(internal, external)

# prediction
pred = model.predict(internal)

# metrics
mae = mean_absolute_error(external, pred)
mse = mean_squared_error(external, pred)
rmse = np.sqrt(mse)
r2 = r2_score(external, pred)

print("Coefficient:", model.coef_[0])
print("Intercept:", model.intercept_)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)

# line fit graph
plt.scatter(internal, external)
plt.plot(internal, pred)
plt.xlabel("Internal Marks")
plt.ylabel("External Marks")
plt.title("Simple Linear Regression")
plt.show()
```
